#####  What is Hypothesis Testing? Why is it used? Importance of Hypothesis Testing.

Hypothesis testing is a statistical method used to make decisions or inferences about a population parameter based on sample data. It involves:

>Stating a null hypothesis (H0) and an alternative hypothesis (H1).

>Determining if there is enough evidence in the sample to reject H0 in favor of H1.

Importance:

>Helps validate research claims.
>Provides a structured framework for decision-making.

##### Create a hypothesis about the average weight of male and female college students.

Hypothesis:

>(Null Hypothesis): The average weight of male college students is equal to the average weight of female college students.

>(Alternative Hypothesis): The average weight of male college students is greater than the average weight of female college students.

##### Python script to conduct a hypothesis test on the difference between two population means.

```python
import scipy.stats as stats

def hypothesis_test_two_means(sample1, sample2):
    t_stat, p_value = stats.ttest_ind(sample1, sample2)
    return t_stat, p_value

sample1 = [80, 85, 90, 95, 100] 
sample2 = [70, 75, 80, 85, 90]   
t_stat, p_value = hypothesis_test_two_means(sample1, sample2)
print(f"T-statistic: {t_stat}, P-value: {p_value}")
```

 ##### What is a null and alternative hypothesis? Examples.

>Null Hypothesis (H0): Assumes no effect or no difference. It is the hypothesis to be tested.

>Alternative Hypothesis (H1): Assumes there is an effect or difference.

Examples:

>H0: There is no difference in test scores between two teaching methods.

>H1: There is a difference in test scores between two teaching methods.

##### Steps involved in Hypothesis Testing.

>State the null and alternative hypotheses.

>Choose a significance level (e.g., 0.05).

>Calculate the test statistic.

>Find the p-value.

>Compare the p-value with the significance level.

>Make a decision to reject or fail to reject the null hypothesis.

#####  Define p-value and its significance.

The p-value is the probability of observing data as extreme as the current sample, assuming the null hypothesis is true. It helps determine whether the results are statistically significant.

>A small p-value (e.g., 
p<0.05) indicates strong evidence against the null hypothesis.

##### A coffee shop owner wants to estimate the average daily revenue for their shop. They take a random sample of 50 days and find the sample mean revenue to be 500 with a standard deviation of 50. Estimate the population mean revenue with a 95% confidence interval.

```python
import scipy.stats as stats

# Given values
sample_mean = 500
std_dev = 50
n = 50
confidence_level = 0.95

# Calculate standard error
standard_error = std_dev / (n ** 0.5)

# Calculate confidence interval
confidence_interval = stats.norm.interval(confidence_level, loc=sample_mean, scale=standard_error)
print(f"95% Confidence Interval: {confidence_interval}")
```

##### A researcher hypothesizes that a new drug will decrease blood pressure by 10 mmHg. They conduct a clinical trial with 100 patients and find that the sample mean decrease in blood pressure is 8 mmHg with a standard deviation of 3 mmHg. Test the hypothesis with a significance level of 0.05.

```python
import scipy.stats as stats

# Given values
mu = 10
sample_mean = 8
std_dev = 3
n = 100
alpha = 0.05

# Calculate t-statistic
t_stat = (sample_mean - mu) / (std_dev / (n ** 0.5))

# Calculate critical value for two-tailed test
t_critical = stats.t.ppf(1 - alpha, df=n-1)

print(f"T-statistic: {t_stat}, Critical value: {t_critical}")
```

##### An electronics company produces a certain type of product with a mean weight of 5 pounds and a standard deviation of 0.5 pounds. A random sample of 25 products is taken, and the sample mean weight is found to be 4.8 pounds. Test the hypothesis that the true mean weight of the products is less than 5 pounds with a significance level of 0.01.

```python
# Given values
mu = 5
sample_mean = 4.8
std_dev = 0.5
n = 25
alpha = 0.01

# Calculate t-statistic
t_stat = (sample_mean - mu) / (std_dev / (n ** 0.5))

# Calculate critical value for one-tailed test
t_critical = stats.t.ppf(alpha, df=n-1)

print(f"T-statistic: {t_stat}, Critical value: {t_critical}")
```

##### A marketing company wants to estimate the average number of ads watched by viewers during a TV program. They take a random sample of 50 viewers and find that the sample mean is 4 with a standard deviation of 1.5. Estimate the population mean with a 99% confidence interval.

```python
import scipy.stats as stats

# Given values
sample_mean = 4
std_dev = 1.5
n = 50
confidence_level = 0.99

# Calculate the Z-score for a 99% confidence interval
z_score = stats.norm.ppf((1 + confidence_level) / 2)

# Calculate the standard error
standard_error = std_dev / (n ** 0.5)

# Calculate the confidence interval
margin_of_error = z_score * standard_error
confidence_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)

print(f"99% Confidence Interval: {confidence_interval}")
```
